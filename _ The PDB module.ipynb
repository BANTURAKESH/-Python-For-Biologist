# Reading a PDB file

```python
from Bio.PDB.PDBParser import PDBParser
Parser = PDBParser(PERMISSIVE=1)
Parser
```

```python
structure_id = "5il7"
```

```python
filename ='5il7.pdb'
```

```python
structure = Parser.get_structure(structure_id,filename)
```

```python
structure
```

```python
filename
```

```python
resolution = structure.header['resolution']
```

```python
keywords = structure.header['keywords']
```
